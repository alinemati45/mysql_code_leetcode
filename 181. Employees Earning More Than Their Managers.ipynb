The `Employee_new` table holds all employees including their managers. Every employee has an Id, and there is also a column for the manager Id.


| Id | Name  | Salary | ManagerId |
|----|-------|--------|-----------|
| 1  | Joe   | 70000  | 3         |
| 2  | Henry | 80000  | 4         |
| 3  | Sam   | 60000  | NULL      |
| 4  | Max   | 90000  | NULL      |



Given the `Employee_new` table, write a SQL query that finds out employees who earn more than their managers. For the above table, Joe is the only employee who earns more than his manager.



| Employee |
|----------|
| Joe      |




```mysql
Create table If Not Exists Employee_new (Id int, Name varchar(255), Salary int, ManagerId int);
Truncate table Employee_new;

```

````python
query("""
select
    a.Name AS 'Employee'
FROM
    Employee_new AS a,
    Employee_new AS b
WHERE
    a.ManagerId = b.Id
        AND a.Salary > b.Salary;
""")
````

````python
query("""

SELECT
     a.NAME AS 'Employee'
FROM Employee_new AS a JOIN Employee_new AS b
     ON a.ManagerId = b.Id
     AND a.Salary > b.Salary;

""")
````
