Write an SQL query to find the type of weather in each country for November 2019.

The type of weather is `Cold` if the average weather_state is less than or equal 15, `Hot` if the average weather_state is greater than or equal 25 and `Warm` otherwise.

Return result table in any order.

````python
query("""
    SELECT AVG(W.weather_state) AS weather_type
    
       FROM Weather as W
        left join  Countries as C on W.country_id = C.country_id 
        where W.day between '2019-11-01' and '2019-11-30'
        GROUP BY country_name

;""")
````

````python
query("""
    SELECT  country_name ,
    CASE 
        WHEN avg_weather<=15 then 'Cold' 
        WHEN avg_weather>=25 then 'Hot' 
        ELSE 'Warm' 
    END as weather_type 
    FROM
    (SELECT AVG(W.weather_state) AS avg_weather , country_name
    FROM Weather as W
        left join  Countries as C on W.country_id = C.country_id 
        where date_format(day,'%Y-%m') = '2019-11'
        GROUP BY country_name ) TMPT

;""")
````
