````python
query("""


SELECT
    e.Name, b.Bonus
FROM
    Employee as e
        LEFT OUTER JOIN
    Bonus as b ON e.Empid = b.Empid 
    WHERE   b.Bonus < 1000 or b.Bonus is null
    order by e.Name;
    
""")
````

````python
query("""

SELECT
    Employee.name, Bonus.bonus
FROM
    Employee
        LEFT OUTER JOIN
    Bonus ON Employee.empid = Bonus.empid 
    WHERE   Bonus.bonus < 1000 or Bonus.bonus is null
    order by Employee.name
    ;

""")
````

````python
query('''

select
    a.name,
    b.bonus 
from
    Employee a 
left outer join
    Bonus b 
on (a.empid = b.empid)
where ifnull(bonus,-1) < 1000

order by a.name
''')
````
