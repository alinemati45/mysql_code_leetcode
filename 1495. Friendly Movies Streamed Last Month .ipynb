Table: TVProgram


| Column Name   | Type    |
|---------------|---------|
| program_date  | date    |
| content_id    | int     |
| channel       | varchar |


(program_date, content_id) is the primary key for this table.
This table contains information of the programs on the TV.
content_id is the id of the program in some channel on the TV.

Table: Content


| Column Name      | Type    |
|------------------|---------|
| content_id       | varchar |
| title            | varchar |
| Kids_content     | enum    |
| content_type     | varchar |

content_id is the primary key for this table.
Kids_content is an enum that takes one of the values ('Y', 'N') where: 
'Y' means is content for kids otherwise 'N' is not content for kids.
content_type is the category of the content as movies, series, etc.

Write an SQL query to report the distinct titles of the kid-friendly movies streamed in June 2020

````python
query("""
SELECT * 
FROM TVProgram as t
left join Content as c on t.content_id = c.content_id
""")
````

````python
query("""
SELECT * 
FROM TVProgram as t
left join Content as c on t.content_id = c.content_id 
where t.program_date between "2020-06-1" and "2020-06-30" 
""")
````

````python
query("""
SELECT * 
FROM TVProgram as t
left join Content as c on t.content_id = c.content_id 
where t.program_date between "2020-06-1" and "2020-06-30"  and kids_content ='Y'
""")
````

````python
query("""
SELECT distinct title
FROM TVProgram as t
left join Content as c on t.content_id = c.content_id 
where t.program_date between "2020-06-1" and "2020-06-30"  and kids_content ='Y'
""")
````

````python
query('''
select distinct title
from Content c, TVProgram tv
where c.content_id = tv.content_id
        and Kids_content = 'Y'
        and content_type = 'Movies'
        and program_date between '2020-06-01' and '2020-06-30'


''')
````

````python
query('''

SELECT DISTINCT title
FROM Content ctt
INNER JOIN TVProgram tv
ON ctt.content_id = tv.content_id
WHERE content_type = 'Movies'
AND Kids_content = 'Y'
AND program_date BETWEEN '2020-06-01' AND '2020-06-30';

''')
````
